## 텐서와 자동 미분

$y=a*x$ ,  $L=\sum_{i} y_i$

$\frac{\delta L}{δa_k}=\sum_{i} x_{ik}$

```
x = torch.randn(100, 3)
# 미분의 변수로 사용하는 경우는 requires_grad를 True로 설정
a = torch.tensor([1, 2, 3.], requires_grad=True)

# 계산을 통해 자동으로 계산 그래프가 구축된다
y = torch.mv(x, a)
l = y.sum()

# 미분을 실행
l.backward()
# 분석 답과 비교
a.grad != x.sum(0)
```

```
print(a.grad)
print(x.sum(0))
```
